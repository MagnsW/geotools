```python
from geotools import offset_dist
```

```python
path = './testfiles/nearoffset_dist/'
filenames_1seq = {'Dual Source Wide on Top': path + '04_Dual_wide_on_top_1seq.A1X',
            'Triple Source Wide in Front': path + '04_Triple_wide_1seq.A1X',
            'Triple Source Wide in Front, U-shaped front': path + '04_Triple_wide_U_1seq.A1X'}
```

```python
my_offset_data_1seq = offset_dist.OffsetData(filenames_1seq)
```

```python
my_offset_data_1seq.plot_swarm(offsetlim=600, xtickinc=4)
```

```python
my_offset_data_1seq.makeoffsetclass(0, 600, 25)
```

```python
my_offset_data_1seq.hist_offsetclass()
```

```python
my_offset_data_1seq.offsetclassbinning()
```

```python
my_offset_data_1seq.plot_offset_bins(maxfold=17)
```
