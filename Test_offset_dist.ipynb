```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
dwot = my_offset_data_5seq.df[(my_offset_data_5seq.df['Configuration'] == 'Dual Source Wide on Top')]
```

```python
dwot
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(16, 12))
sns.swarmplot(x='MidPtX', y='AzSrc', data=dwot[(dwot['Offset'] <= 600)])
#plt.ylim(600, 0)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(16, 12))
sns.swarmplot(x='MidPtX', y='AzSrc', data=my_offset_data_5seq.df[(my_offset_data_5seq.df['Offset'] <= 600)], hue='Configuration')
#plt.ylim(600, 0)
plt.show()
```
